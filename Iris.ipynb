```python
#import liba
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading dataset

```python
columns =['SepalLengthCm'	,'SepalWidthCm','	PetalLengthCm',	'PetalWidthCm',	'Species']

#load the data
df = pd.read_csv('Iris.csv', names=columns)
df.head() # give us the top five elements of the data
```

#visualization of data

```python
df.describe()
```

```python

# Create the pairplot
sns.pairplot(df, hue='Species')
```

#separate features and target

```python
#separate features and target
data =df.values
x = data [:,0:4]
y = data [:,4]
print (y)
```


#split the data into trainng and testing


```python


 from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print (x_train)
```

# Model 1 - support vector machine algorithm

```python

from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```

```python
prediction1 = model.predict(x_test)
#calc the accuracy

from sklearn.metrics import accuracy_score
print (accuracy_score(y_test,prediction1)*100)


''' to see the 3% of error
we can do a for loop  to
print the comparsion
like ->
from i in range (len(prediction1)):
  print (prediction1[i], y_test[i])
'''
```

#Model 2 - logistic regression

```python
 from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
model2.fit(x_train, y_train)
```

```python
prediction2 = model2.predict(x_test)
#calc the accuracy

from sklearn.metrics import accuracy_score
print (accuracy_score(y_test,prediction2)*100)
#
```
